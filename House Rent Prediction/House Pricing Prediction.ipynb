```python
def missing_data(df,n):
    missing = df.isna().sum().sort_values(ascending=False)
    percentage = (df.isna().sum()/df.isnull().count()) *100
    unique = df.nunique()
    
    missing_values = pd.concat([missing,percentage,unique],axis=1,keys=["Total_missing","percentage","Unique_values"])
    
    print(missing_values.head(n))
    
```

```python
missing_data(df_new,20)
```

```python
def filter1(df,x):
    percent_of_missing = x
    threshold = (1 - (percent_of_missing/100))* len(df)
    df.dropna(axis=1,thresh=threshold,inplace = True)
    
    print(df.head(x))
```

```python
filter1(df_new,70)
```
